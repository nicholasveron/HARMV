````python
def repeat_to_length(s, wanted):
    return (s * (wanted//len(s) + 1))[:wanted]

def str_add_to_hex(str_a , str_b):
    maxlen = max(len(str_a),len(str_b))
    str_a = repeat_to_length(str_a,maxlen)
    str_b = repeat_to_length(str_b,maxlen)
    l_str_a = list(str_a)
    for i , ch in enumerate(str_b):
        l_str_a[i] = chr(
            (
                (ord(l_str_a[i])) + (ord(ch))
            )
        )
    return ("".join(l_str_a)).encode()

def str_sub_hex_w_str(hex_a , str_b):
    str_a = hex_a.decode()
    maxlen = max(len(str_a),len(str_b))
    str_a = repeat_to_length(str_a,maxlen)
    str_b = repeat_to_length(str_b,maxlen)
    l_str_a = list(str_a)
    for i , ch in enumerate(str_b):
        l_str_a[i] = chr(
            (
                (ord(l_str_a[i])) - (ord(ch))
            )
        )
    return "".join(l_str_a)
````
